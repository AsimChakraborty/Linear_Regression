```python
df1=newdf.drop(columns='age',axis=1)
```

```python
x=df1.drop(['cardio','id'],axis=1)
```

```python
x
```

```python
y=df1['cardio']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.30,random_state=2)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier(n_estimators=100)
```

```python
rc.fit(xtrain,ytrain)
```

```python
rc.score(xtest,ytest)
```

#DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier()
```

```python
dc.fit(xtrain,ytrain)
```

```python
dc.score(xtest,ytest)
```

#KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier()
```

```python
kn.fit(xtrain,ytrain)
```

```python
kn.score(xtest,ytest)
```

#svm

```python
from sklearn.svm import SVC
sc=SVC()
```

```python
sc.fit(xtrain,ytrain)
```

```python
sc.score(xtest,ytest)
```

#Naive Bayes.

```python
from sklearn.naive_bayes import GaussianNB
nv=GaussianNB()
```

```python
nv.fit(xtrain,ytrain)
```

```python
nv.score(xtest,ytest)
```

#prediction 

```python
prediction=nv.predict([[1,156,85.0,140,90,3.0,1.0,0.0,0.0,1.0,55.0]])
print(prediction)
print("*******************")
if prediction ==0:
  print("No Heart Disease")
else:
  print("Heart Disease")  
```
